```python
import pandas as pd
df = pd.read_csv("Data set\weather_data.csv", parse_dates=["day"])
df.set_index('day', inplace = True)
df
```

```python
df.isnull().sum()
```

```python
new_df=df.fillna(0)
new_df
```

# fill the missing values with fill na method with diferent type.

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': "No event"
})
new_df
```

# "ffill" method in fillna method is fullfill the none values with previous data values

```python
new_df = df.fillna(method="ffill")
new_df
```

# "bfill" method in fillna method is fullfill the none values with next data values

```python
new_df = df.fillna(method="bfill")
new_df
```

# "bfill,  axis = 'columns' " method in fillna method is fullfill the none values with data values in next column

```python
new_df = df.fillna(method="bfill", axis= "columns")
new_df
```

# The limit parameter in fillna with method='ffill' fills only up to one consecutive missing value, limiting forward-fill operations.


```python
new_df = df.fillna(method='ffill',limit=1)
new_df
```

```python
new_df = df.fillna(method='ffill',limit=2)
new_df
```

# Using interpolate()method to get best values to fulfill missing values.

```python
new_df = df.interpolate()
new_df
```

#  Using interpolate( method ="time")method to get powerful best values to fulfill missing values.

```python
new_df = df.interpolate(method='time')
new_df
```

# .dropna() method to drop the all missing values

```python
new_df = df.dropna()
new_df
```

# If you need frop the missing value row , which all column's values are missing, use below method 

```python
new_df = df.dropna(how='all')
new_df
```

# if there is no atleast one valid values use this methode to drop it

```python
new_df = df.dropna(thresh=1)
new_df
```

# Need to add new indexes to current dataframe with missing values according to current indexes range

# Handling special values in a dataframe

```python
import pandas as pd
import numpy as np
df = pd.read_csv("Data set\weather_data2.csv")
df
```

```python
new_df = df.replace(-99999, np.nan)
new_df
```

# Neet to replace values based on columns

```python
new_df = df.replace({
    'temperatue': -99999,
    'windspeed': -99999,
    'event': 0
},np.nan)

new_df
```

```python
df = pd.read_csv("Data set\weather_data2.csv")
new_df = df.replace({
    -99999: np.nan,
    'No event' : 'Sunny'
})
new_df
```

# How "regex" Regular Expression is working

```python
df = pd.read_csv("Data set\weather_data2.csv")
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'', regex=True)
new_df
```
